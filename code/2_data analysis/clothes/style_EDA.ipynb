```python
# _소재 멀티라벨 인코딩
list_cols = [col for col in data_drop.columns if col.endswith("_소재")]
print(list_cols)
```

```python
data_drop['상의_소재'].value_counts()
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
import ast
# -------------------------
# 2) 리스트 변환 함수 (문자열 리스트도 처리)
# -------------------------
def to_list(x):
    # 1) 결측
    if pd.isna(x):
        return ["nan"]

    # 2) 이미 리스트
    if isinstance(x, list):
        return [str(v).strip() for v in x if (v is not None and str(v).strip()!="")]

    # 3) 문자열: 여러 케이스 처리
    if isinstance(x, str):
        s = x.strip()
        # a) Python list 문자열인 경우: "['a','b']" or '["a","b"]'
        try:
            val = ast.literal_eval(s)
            if isinstance(val, list):
                return [str(v).strip() for v in val if (v is not None and str(v).strip()!="")]
        except Exception:
            pass
        # b) 쉼표로 구분된 경우: "a, b, c" -> split
        if "," in s:
            parts = [p.strip() for p in s.split(",") if p.strip()!=""]
            return parts
        # c) 구분자가 '|' 또는 '/' 등일 경우 추가 처리
        for sep in ["|", "/", ";"]:
            if sep in s:
                parts = [p.strip() for p in s.split(sep) if p.strip()!=""]
                return parts
        # d) 그 외 단일값
        return [s]

    # 4) 숫자 등 기타
    return [str(x).strip()]
```

```python

# -------------------------
# 3) MultiLabelBinarizer 인코딩 적용
# -------------------------
for col in list_cols:
    # 실제 리스트로 변환
    data_drop[col] = data_drop[col].apply(to_list)
    
    # MultiLabelBinarizer 적용
    mlb = MultiLabelBinarizer()

    expanded = pd.DataFrame(
        mlb.fit_transform(data_drop[col]),
        columns=[f"{col}_{cls}" for cls in mlb.classes_],
        index=data_drop.index
    )

    # 원본 컬럼 삭제 + 확장 컬럼 합치기
    data_drop = pd.concat([data_drop.drop(columns=[col]), expanded], axis=1)

# -------------------------
# 끝: data_drop는 모든 _소재 컬럼이 multi-hot 인코딩된 상태
# -------------------------

print(data_drop.head())
```
