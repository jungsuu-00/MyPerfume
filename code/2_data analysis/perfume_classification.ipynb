```python
# Cluster → 설문조사 연결 매핑
cluster_mapping = {
    '0': "파우더느낌의 부드러운향",
    '1': "플로럴향, 달콤한향",
    '2': "싱그러운 풀 향",
    '3': "머스크같은 중후한향",
    '4': "플로럴향, 달콤한향",
    '5': "파우더느낌의 부드러운향",
    '6': "싱그러운 풀 향",
    '7': "파우더느낌의 부드러운향",
    '8': "플로럴향, 달콤한향",
    '9': "라벤더같은 은은한 향",
    '10': "감귤류의 상큼한 향"
}

# 라벨링 적용
df_cluster["cluster"] = df_cluster["cluster"].astype(str)
df_cluster["향분류"] = df_cluster["cluster"].map(cluster_mapping)

df_cluster = df_cluster.iloc[:,[0,-1]]
```

```python
df_cluster['향분류'].value_counts()
```

```python
# 1. perfume_id 기준으로 병합
df_merged = pd.merge(df, df_cluster, on="perfume_id", how="inner")

# 2. 동일 여부 컬럼 생성 (True/False)
df_merged["match"] = df_merged["fragrance"] == df_merged["향분류"]

# 3. 동일 비율 계산
match_ratio = df_merged["match"].mean()  # True=1, False=0 → 평균 = 비율

print("대분류와 향분류 동일 비율:", match_ratio)
```
